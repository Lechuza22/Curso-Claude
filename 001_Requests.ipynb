```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-opus-4-7"
```

```python
# Make request

message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "Que es la computación cuantica, responde en una oración"
        }
    ]
)
print(message.content[0].text)
```

```python
# Para mantener el contexto y que las respuestas sean sobre el mismo tema. 


def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
```

```python
# utilizar las funciones para mantener la conversacion. 
# Start with an empty message list
messages = []

# Add the initial user question
add_user_message(messages, "Que es la computación cuantica, responde en una oración")

# Get Claude's response
answer = chat(messages)

# Add Claude's response to the conversation history
add_assistant_message(messages, answer)

# Add a follow-up question
add_user_message(messages, "Escribir otra oración")

# Get the follow-up response with full context
final_answer = chat(messages)

messages
```

```python
#  indicaciones al sistema 

system_prompt = """
Tu eres un paciente tutor de matematicas, 
no daras la respuesta al estudiante que pregunta, 
y lo guirás a la sulución paso por paso.
"""

client.messages.create(
    model=model,
    messages=messages,
    max_tokens=1000,
    system=system_prompt
)
```

```python
def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }
    
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text
print(message.content[0].text)
```

```python
# Without system prompt
answer = chat(messages)

# With system prompt
system = """
Tu eres un paciente tutor de matematicas, 
no daras la respuesta al estudiante que pregunta, 
y lo guirás a la sulución paso por paso.
"""
answer = chat(messages, system=system)
print(message.content[0].text)
```
