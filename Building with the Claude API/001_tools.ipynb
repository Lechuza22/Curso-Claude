```python
# Load env variables and create client
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
# Helper functions
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages, system=None, temperature=1.0, stop_sequences=[]):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
        "stop_sequences": stop_sequences,
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
from datetime import datetime

def get_current_datetime():
    return datetime.now().strftime("%H:%M:%S")

get_current_datetime_schema = {
    "name": "get_current_datetime",
    "description": "Returns the current time in HH:MM:SS format.",
    "input_schema": {
        "type": "object",
        "properties": {}
    }
}

# Paso 1: Claude decide usar el tool
messages = [{"role": "user", "content": "What is the exact time, formatted as HH:MM:SS?"}]

response = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    tools=[get_current_datetime_schema],
)

# Paso 2: ejecutar la función y devolver el resultado
if response.stop_reason == "tool_use":
    tool_use = next(b for b in response.content if b.type == "tool_use")
    result = get_current_datetime()  # ejecutamos la función real

    messages.append({"role": "assistant", "content": response.content})
    messages.append({
        "role": "user",
        "content": [{
            "type": "tool_result",
            "tool_use_id": tool_use.id,
            "content": result
        }]
    })

    # Paso 3: Claude responde con el resultado
    final = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
        tools=[get_current_datetime_schema],
    )
    print(final.content[0].text)
```

```python
from datetime import datetime

def get_current_datetime():
    return datetime.now().strftime("%H:%M:%S")

get_current_datetime_schema = {
    "name": "get_current_datetime",
    "description": "Returns the current time in HH:MM:SS format.",
    "input_schema": {"type": "object", "properties": {}}
}

# Paso 1: enviar mensaje de usuario con el schema del tool
messages = [{"role": "user", "content": "What is the exact time, formatted as HH:MM:SS?"}]

response = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    tools=[get_current_datetime_schema],
)

# Paso 2: recibir respuesta del asistente (texto + bloque tool_use)
tool_use = next(b for b in response.content if b.type == "tool_use")

# Paso 3: extraer info del tool y ejecutar la función
result = get_current_datetime()

# Paso 4: agregar historial completo + resultado del tool
messages.append({
    "role": "assistant",
    "content": response.content       # ← incluye texto y bloque tool_use
})
messages.append({
    "role": "user",
    "content": [{
        "type": "tool_result",
        "tool_use_id": tool_use.id,
        "content": result
    }]
})

# Paso 5: recibir respuesta final de Claude
final = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    tools=[get_current_datetime_schema],
)
print(final.content[0].text)
```
